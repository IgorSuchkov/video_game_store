Помимо значений, есть и пропуски nan. Для преобразования столбца нужно заполнить пропуски. Посмотрим на пропуски.

В столбце year_of_release часть пропусков можно восстановить. Причины пропуска могут быть в том, что при заполнениеии забыли указать. Восстанавливать будем по названию игры, так как в некоторых название всречается год игры, который обозначает дату релиза.

```python
#Пропуски в дате выхода игр
isna_year_of_released = df[df['year_of_release'].isna()]
isna_year_of_released
```

отсортируем данные по дате релиза из известных нам, чтобы исключить игры, которые обозначают не дату релиза, а например время в котором происходит действие. И возьмем из него минимальное и максимальное значение.

```python
#сортировка данных
sorted(df.year_of_release.unique())
```

Поставим значение даты релиза в играх, где встречаются числа больше 1980 и меньше 2016.

```python
#создаем цикл для заполнения значений
for i in isna_year_of_released.name:
    for j in i.split():
        if j.isdigit() == True:
            if int(j) >= 1980 and int(j) <= 2016:
                if j in i:
                    df.loc[df['year_of_release'].isna() & df['name'] == i, 'year_of_release'] = int(j)
```

Посмотрим, как еще могут обозначаться года релиза в названиях игр.

```python
df[df['year_of_release'].isna()].name.unique()
```

Были найденые и другие абривиатуры, которые обозначают название игр.

Добавим значения в столбце релиза оп другим абривиатура, сначало по численным.

```python
#создаем цикл для заполнения значений
for i in isna_year_of_released.name:
    for j in i.split():
        if j.isdigit() == True:
            if int(j) == 99:
                if j in i:
                    df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 1999
            elif j == '07':
                if j in i:
                    df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2007
            elif int(j) == 11:
                if j in i:
                    df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2011
            elif j == '06':
                if j in i:
                    df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2006
```

И по строковым значениям.

```python
#создаем цикл для заполнения значений
for i in isna_year_of_released.name:
    for j in i.split():
        if j == '20-03':
            if j in i:
                df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2003
        elif j == '(2013)':
            if j in i:
                df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2013
        elif j == '2K6':
            if j in i:
                df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2006
        elif j == '2K8':
            if j in i:
                df.loc[df['year_of_release'].isna() & (df['name'] == i), 'year_of_release'] = 2008
```

```python
#процент пропусков
df.isna().sum().sort_values()/df.shape[0]*100
```

В столбце year_of_release пропущенных данных не много, около 1 процента. Заполнять другими значениями нет смысла, чтобы не получились аномали по годам релиза. Поэтому по совокупности фактора просто удалим их.

```python
# Удаляем лишнии данные.
df = df.dropna(subset=['year_of_release']).reset_index(drop=True)
```

Остальные пропуски заполним таким образом. Из тех данных что есть сгруппируем по платформам, а затем внутри платформ по жанрам, в каждом жанре найдем моду, и ей заполним пропуски в этом в каждой платформе по жанрам платформы.

```python
#перебираем данные по платформе
for i in df.platform.unique():
    #выделяем данные по платформе
    rt_pl = df.query('platform == @i')
    #передираем все жанры в ново датафрейме
    for j in rt_pl.genre.unique():
        #внутри этих данных выделяем данные по жанру
        rt_gn = df.query('genre == @j')
        #присваимваем новое значение по моде
        df.loc[(df.platform == i) & (df.genre == j) & df.rating.isna(), 'rating'] = rt_gn.rating.mode()[0]
```

```python
#создание словарь с полявением платформ
dic = {}
for i in df_group.index:
    if i[1] in dic:
        if i[0] < dic.get(i[1]):
            dic[i[1]] = i[0]
    else:
        dic[i[1]] = i[0]
```

```python
#разделим года на количество получившихся платформ.
agr = 0
for i in  range(len(dic.values())):
    if i != (len(dic.values())-1):
        add = (list(dic.values())[i+1] - list(dic.values())[i])
        agr += add
print('Платформ ежегодно появляется около', round(agr/len(dic.values())))
```

## Проверка гипотез.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем 0 гипотезу.

```python
f'Количество пропусков {df.user_score.isna().sum()} из {df.shape[0]}'
```

Так как часть данных отсутствуют, то имеем дело, не генератльной совокупностью, а только с ее выборкой. Поэтому нулевую гипотезу формулируем так. Если средний пользовательский рейтинг выборки XOne равен среднему пользовательскую рейтингу выборки платформы PC, то и среднии их генеральных совокупностей равны.

```python
print('количество значений в выборке рейтинга XOne')
display(df.loc[df['platform'] == 'XOne']['user_score'].count())
print('количество значений в выборке рейтинга PC')
df.loc[df['platform'] == 'PC']['user_score'].count()
```

Сформулированная гипотеза является двусторонней. Так как это выборка из генеральной совокупности и количество значений выборки больше 30, то мы применяем формулу для стандартной ошибки среднего SE, а в коде ttest_ind. 


Альтернативная гипотеза заключается в том, что если значению средних наших выборок не поподает в выбранный нами доверительный интервал, то и среднии генеральной совокупности не равны, и в этом случаи отклоняется нулевая гипотеза.

H_0: Рейтинг пользователей (user_score) платформы "XOne" = Рейтинг пользователей (user_score) платформы "PC"

H_1: Рейтинг пользователей (user_score) платформы "XOne" ≠ Рейтинг пользователей (user_score) платформы "PC"

Проверка гипотезы.

Для сравнение двух платформ необходимо ограчничить выборку по платформе PC, так как платформа выпускается 30 лет и люди по разному оценивила игры, по сравнению  с более стабильным периодом, когда вышла другая платформ. Чтобы эта разница не повлияла на результаты исследования ограничи данные по рейтингу для платформы PC с начала появления другой плаформы.

```python
#находим дату релиза платформы
year_XOne = df.loc[df['platform'] == 'XOne']['year_of_release'].min()
```

```python
#возьмем данные после появления XOne.
df_for_rating_PC = df.query('year_of_release >= @year_XOne')
```

```python
#доверительный интервал
alpha = 0.05

#Выборки
Rating_XOne = list(df.loc[df['platform'] == 'XOne']['user_score'].dropna())
Rating_PC = list(df_for_rating_PC.loc[df['platform'] == 'PC']['user_score'].dropna())

#проверка гипотезы
results = st.ttest_ind(Rating_XOne, Rating_PC, equal_var = False)
print(results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Данные пользовательских рейтингов у платформ XOne и PC не противоречат нулевой гипотезе.

Означает, что не получилось отвергнуть 0 гипотезу, а это значит, что данные не противоречат гипотезе.

### Средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем 0 гипотезу.

```python
f'Количество пропусков {df.genre.isna().sum()} из {df.shape[0]}'
```

Мы имеем дело с генеральной совокупностью, далее ГС. Поэтому нулевую гипотезу формулируем так. Средний пользовательский рейтинг ГС жанра Action равен среднему пользовательскому рейтингу ГС жанра Sports.

Сформулированная гипотеза является двусторонней о равенстве средних двух выборок.

Альтернативная гипотеза заключается в том, что если значение средней ГС не поподает в выбранный нами доверительный интервал, то среднии ГС не равны, и в этом случаи отклоняется нулевая гипотеза.

H_0: Рейтинг пользователей (user_score) жанра "Action" = Рейтинг пользователей (user_score) жанра "Sports"

H_1: Рейтинг пользователей (user_score) жанра "Action" ≠ Рейтинг пользователей (user_score) жанра "Sports"

Проверка гипотезы.

```python
#доверительный интервал
alpha = 0.05

#Выборки
Rating_Action = list(df.loc[df['genre'] == 'Action']['user_score'].dropna())
Rating_Sports = list(df.loc[df['genre'] == 'Sports']['user_score'].dropna())

#проверка гипотезы
results = st.ttest_ind(Rating_Action, Rating_Sports, equal_var = False)
print(results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Данные пользовательских рейтингов у жанров Action и Sports не противоречат нулевой гипотезе.

**Вывод:** Мы провели проверки статистический гипотез и удалось выяснить, что средние пользовательские рейтинги платформ XOne и PC не противоречат 0 гипотезе, и средние пользовательские рейтинги жанров Action и Sports не противоречат гипотезе о равенстве средних.
